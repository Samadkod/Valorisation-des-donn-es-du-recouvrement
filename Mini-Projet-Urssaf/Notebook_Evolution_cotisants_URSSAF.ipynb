
### 沒 Construction du graphique

L'objectif ici est de visualiser l'ﾃｩvolution du nombre de cotisants par rﾃｩgion.

汨 **Comment lire le graphique ?**
- Chaque courbe reprﾃｩsente une rﾃｩgion.
- L'axe horizontal montre les annﾃｩes.
- L'axe vertical montre le nombre de cotisants.

On cherche ﾃ rendre la lecture immﾃｩdiate pour un public non tecnhique (direction mﾃｩtier, partenaires externes).


###  Crﾃｩation d'un graphique illustrant l'ﾃｩvolution du nombre de cotisants URSSAF par rﾃｩgion (2019-2023)

```python

# Crﾃｩation ce graphique

plt.figure(figsize=(10, 6))
for region in df['Rﾃｩgion'].unique():
    region_data = df[df['Rﾃｩgion'] == region]
    plt.plot(region_data['Annﾃｩe'], region_data['Nombre de cotisants'], marker='o', label=region)

plt.title("ﾃ益olution rﾃｩgionale du nombre de cotisants URSSAF (2019-2023)")
plt.xlabel("Annﾃｩe")
plt.ylabel("Nombre de cotisants")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### En somme 

Ce graphique simple permet de :
- Mettre en ﾃｩvidence la dynamique des cotisants par rﾃｩgion
- Aider les ﾃｩquipes mﾃｩtier ﾃ suivre les ﾃｩvolutions rﾃｩcentes
- Faciliter le dialogue avec les partenaires sur les enjeux ﾃｩconomiques et sociaux du recouvrement.

汨 Cette analyse peut ﾃｪtre facilement intﾃｩgrﾃｩe dans un reporting automatisﾃｩ pﾃｩriodique pour les ﾃｩquipes de pilotage et les partenaires institutionnels.

```python
# Crﾃｩons une table montant le taux de croissance dans le temps par rﾃｩgion
variation_table = []

# Boucle sur chaque rﾃｩgion
for region in df['Rﾃｩgion'].unique():
    # Valeur 2020
    value_2020 = df[(df['Rﾃｩgion'] == region) & (df['Annﾃｩe'] == 2020)]['Nombre de cotisants'].values[0]
    # Valeur 2023
    value_2023 = df[(df['Rﾃｩgion'] == region) & (df['Annﾃｩe'] == 2023)]['Nombre de cotisants'].values[0]
    # Calcul variation %
    variation_pct = ((value_2023 - value_2020) / value_2020) * 100
    
    # Ajout ﾃ la table
    variation_table.append({
        'Rﾃｩgion': region,
        'Cotisants 2020': value_2020,
        'Cotisants 2023': value_2023,
        'Variation % 2020-2023': round(variation_pct, 2)
    })

# Conversion en dataframe
variation_df = pd.DataFrame(variation_table)

# Afficher la table finale
variation_df
```
